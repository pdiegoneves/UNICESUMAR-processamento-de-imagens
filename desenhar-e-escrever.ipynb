```python
import numpy as np
import cv2

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
sns.set_style("whitegrid", {"axes.grid": False})

from skimage import io
```

```python
imagem_color = np.zeros((1024, 1024, 3), np.uint8)

plt.imshow(imagem_color)
```

```python
canvas = np.ones((1024, 1024, 3), np.uint8) * 255
plt.imshow(canvas)
```

```python

canvas = np.ones((1024, 1024, 3), np.uint8) * 255
cv2.line(canvas, (0, 0), (1024, 1024), (0, 0, 255), 20)

plt.imshow(canvas)
plt.title("Linha diagonal")
```

```python
canvas = np.ones((1024, 1024, 3), np.uint8) * 255

cv2.line(canvas, (512, 0), (512, 1024), (255, 0, 0), 10)

plt.imshow(canvas)
plt.title("Linha vertical")
```

```python
canvas = np.ones((1024, 1024, 3), np.uint8) * 255

cv2.line(canvas, (0, 512), (1024, 512), (0, 255, 0), 30)

plt.imshow(canvas)
plt.title("Linha horizontal")
```

```python
canvas = np.zeros((1024, 1024, 3), np.uint8)

cv2.rectangle(canvas, (100, 100), (924, 924), (0, 255, 0), 20)

plt.imshow(canvas)
plt.title("Quadrado ")
```

```python
canvas = np.zeros((1024, 1024, 3), np.uint8)

cv2.rectangle(canvas, (100, 100), (512, 512), (255, 0, 0), -1)

plt.imshow(canvas)
plt.title("Quadrad Preenchido")
```

```python
canvas = np.zeros((1024, 1024, 3), np.uint8)

cv2.circle(canvas, (512, 512), 400, (255, 255, 0), 10)

plt.imshow(canvas)
plt.title("Circulo")
```

```python
canvas = np.zeros((1024, 1024, 3), np.uint8)

cv2.circle(canvas, (512, 512), 200, (0, 255, 0), -1)

plt.imshow(canvas)
plt.title("Circulo preenchido")
```

```python
imagem_color = np.zeros((1080, 1920, 3), np.uint8)

cv2.putText(imagem_color, "Imagem", (150, 400), cv2.FONT_HERSHEY_SIMPLEX, 8, (0, 0, 255), 10)

plt.imshow(imagem_color)
plt.title("Texto")
```
