```python
import numpy as np
import cv2

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
sns.set_style("whitegrid", {"axes.grid": False})

from skimage import io
```

```python
imagem = cv2.imread("./img/pexels-photo-12325258.jpeg")
Imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(30, 20))
plt.imshow(imagem)
plt.title("Documento")
```

```python
img = cv2.imread("./img/pexels-photo-12325258.jpeg")
height, width = img.shape[:2]

matriz_rotacao = cv2.getRotationMatrix2D((width, height), -5.4, 1)

out = cv2.warpAffine(img, matriz_rotacao, (width, height))

plt.imshow(out)
plt.title("Imagem rotacionada")
```
